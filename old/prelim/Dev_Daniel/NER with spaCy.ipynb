```python
## word frequency
from collections import Counter

words = [token.text for token in test_article_nlp
         if not token.is_stop and not token.is_punct]

word_freq = Counter(words)

# 5 commonly occurring words with frequencies
common_words = word_freq.most_common(5)
print(common_words)

# unique words
unique_words = [word for (word, freq) in word_freq.items() if freq == 1]
print(unique_words)
print(len(unique_words))
```

```python
## remove stop words from article, retain NOUN, PROPN, ADJ only

%%time 

# set file path
path = '/content/drive/MyDrive/Colab Notebooks/FourthBrain ML Engineer/4B_GLG/Datasets/X_train_tokenized_4e.csv'

# measure elapsed time
from time import process_time


# initialize variables
batch = 1
start = 0
end = start
records = 10
end_specified = 20

# tokenize title
while end <= end_specified + 1:

  # %time for i in range(start,end):
  time_start = process_time()

  for i in range(start,end):
    X_train.article_tokenized.iloc[i] = [token for token in nlp(X_train.article_clean.iloc[i]) if is_token_allowed(token) and token.pos_ in ('NOUN','PROPN','ADJ')]

  start = end

  if end == X_train.shape[0]:
    break
  elif X_train.shape[0] - end < records:
    end = X_train.shape[0]
  else:
    end = start + records 

  time_stop = process_time()
  print("Batch", batch, "elapsed time (min):", (time_stop - time_start)/60)
  batch += 1
  
X_train[0:end].to_csv(path, mode='a', header=False)

print("Done!")
print(start, end)
```
