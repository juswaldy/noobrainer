```python
## preprocessing function that applies:
## lowercasing, lemmatizing, remove punctuation, remove stop words

def is_token_allowed(token):
  if (not token or not token.string.strip() or token.is_stop or token.is_punct):
    return False
  return True

def preprocess_token(token):
  return token.lemma_.strip().lower()
```

# Tokenizing X_train - title_clean

```python
# export X_train (tokenized title), part 1
path = '/content/drive/MyDrive/Colab Notebooks/FourthBrain ML Engineer/4B_GLG/Datasets/NER_data_1a/X_train_tokenized_1a.csv'

# creates the file and adds header info
with open(path, 'w', encoding = 'latin1') as f:
  X_train[0:0].to_csv(f)
```

```python
X_train.shape
```

```python
X_train.head(30)
```

```python
## this finally worked. Commented out for now to continue.

## remove stop words from title, retain NOUN, PROPN, ADJ only

# measure elapsed time
from time import process_time

# initialize variables
batch = 1
start = 0
end = start
records = 1000
end_specified = 45000


# tokenize title
while end <= end_specified + 1:

  print('Start', start, '\n', 'End', end, '\n')

  time_start = process_time()

  for i in range(start,end):
    X_train.title_tokenized.iloc[i] = [token for token in nlp(X_train.title_clean.iloc[i]) if is_token_allowed(token) and token.pos_ in ('NOUN','PROPN','ADJ')]

  start = end

  if end == X_train.shape[0]:
    break
  elif X_train.shape[0] - end < records:
    end = X_train.shape[0]
  else:
    end = start + records 

  time_stop = process_time()
  print("Batch", batch, "elapsed time (min):", (time_stop - time_start)/60)
  batch += 1

  X_train[start:end].to_csv(path, mode='a', header=False)

print("Done!")
```
